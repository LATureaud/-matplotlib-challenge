```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data = "data/mouse_drug_data.csv"
clinical_trial_data = "data/clinicaltrial_data.csv"

mouse_drug_data = pd.read_csv(mouse_drug_data)
clinical_trial_data = pd.read_csv(clinical_trial_data)

# Read the Mouse and Drug Data and the Clinical Trial Data
all_mouse_data =pd.merge(mouse_drug_data, clinical_trial_data, on="Mouse ID", how="left")
all_mouse_data.head()

# Combine the data into a single dataset


# Display the data table for preview

```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_volume = all_mouse_data.groupby(['Drug', 'Timepoint']).agg({'Tumor Volume (mm3)':'mean'})
mean_tumor_volume.reset_index(level=['Drug', 'Timepoint'])
mean_tumor_volume.head()

# timepoint
# tumor_volume
# Drug

# Convert to DataFrame



# Preview DataFrame
```

```python
count_mice = all_mouse_data.groupby(['Drug', 'Timepoint']).mean()
count_mice.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
stand_err_tumor = all_mouse_data.groupby(['Drug','Timepoint']).agg({'Tumor Volume (mm3)':'sem'})
stand_err_tumor.head() 

# stand_err_tumor = all_mouse_data.groupby(['Drug', 'Timepoint']).agg({'Tumor Volume (mm3)':'mean'}).sem
# stand_err_tumor.reset_index(level=['Drug', 'Timepoint']).sem 


```

```python
# Minor Data Munging to Re-Format the Data Frames
#all_mouse_data.columns
df_format= all_mouse_data.pivot_table(index='Timepoint', columns='Drug', values ='Tumor Volume (mm3)')
# Preview that Reformatting worked
df_format.head()
```

```python
# Generate the Plot (with Error Bars)
# Generate the Plot (with Error Bars)
plot_a = df_format.plot (figsize=(9,6), 
            #style=['^-', 'o--', 'x-.', 'D-', 's--', '*--', 'h--', '+--', 'v--','d--'],
            #yerr = df_format,
            markersize=6)
plt.errorbar(df_format.index, df_format['Capomulin'], yerr= df_format['Capomulin'].sem(), marker='x', linestyle='--')
plt.errorbar(df_format.index, df_format['Ceftamin'], yerr= df_format['Ceftamin'].sem(), marker='x', linestyle='--')
plt.errorbar(df_format.index, df_format['Infubinol'], yerr= df_format['Infubinol'].sem(), marker='x', linestyle='--')
plt.errorbar(df_format.index, df_format['Ketapril'], yerr= df_format['Ketapril'].sem(), marker='x', linestyle='--')
plt.errorbar(df_format.index, df_format['Naftisol'], yerr= df_format['Naftisol'].sem(), marker='x', linestyle='--')
plt.errorbar(df_format.index, df_format['Placebo'], yerr= df_format['Placebo'].sem(), marker='x', linestyle='--')
plt.errorbar(df_format.index, df_format['Propriva'], yerr= df_format['Propriva'].sem(), marker='x', linestyle='--')
plt.errorbar(df_format.index, df_format['Ramicane'], yerr= df_format['Ramicane'].sem(), marker='x', linestyle='--')
plt.errorbar(df_format.index, df_format['Stelasyn'], yerr= df_format['Stelasyn'].sem(), marker='x', linestyle='--')
plt.errorbar(df_format.index, df_format['Zoniferol'], yerr= df_format['Zoniferol'].sem(), marker='x', linestyle='--')

#df_format.plot(x ='Time (Days)', y ='Tumor Volume(mm3)')
# Set the xlabel and ylabel using class methods
plot_a.set_xlabel("Time")
plot_a.set_ylabel("Tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")
#plot.legend([Drugs], loc='best')

# Save the Figure

# Save the Figure

```

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

count_mice = all_mouse_data.groupby(['Drug', 'Timepoint']).agg({'Mouse ID':'count'})
count_mice.head()

# Convert to DataFrame

# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames
count_df = all_mouse_data.groupby(['Timepoint', 'Drug']).agg({'Mouse ID':'count'})
count_df.head()
# survival_count = count_mice['Mouse ID'].sum()
# survival_count
# df_count_format = all_mouse_data.pivot_table(index='Timepoint', columns='Drug', values = 'Mouse ID')
# df_count_format.head()
# Preview the Data Frame
# df_met_format = all_mouse_data.pivot_table(index='Timepoint', columns='Drug', values ='Metastatic Sites')
# df_met_format.head()
```

```python
df_test = count_df.reset_index()
df_test = df_test.pivot(index='Timepoint', columns='Drug', values = 'Mouse ID')
df_test

```

```python
# Generate the Plot (Accounting for percentages)

plt.plot(df_test.index.values, df_test.loc[:,"Capomulin"] / df_test.loc[0,"Capomulin"] * 100,
         marker ='o', color='red', label="Capomulin", linewidth=.5)
plt.plot(df_test.index.values, df_test.loc[:,"Ceftamin"] / df_test.loc[0,"Ceftamin"] * 100,
         marker ='x', color='yellow', label="Ceftamin", linewidth=.5)
plt.plot(df_test.index.values, df_test.loc[:,"Infubinol"] / df_test.loc[0,"Infubinol"] * 100,
         marker ='^', color='blue', label="Infubinol", linewidth=.5)
plt.plot(df_test.index.values, df_test.loc[:,"Ketapril"] / df_test.loc[0,"Ketapril"] * 100,
         marker ='s', color='green', label="Ketapril", linewidth=.5)
plt.plot(df_test.index.values, df_test.loc[:,"Naftisol"] / df_test.loc[0,"Naftisol"] * 100,
         marker ='D', color='orange', label="Naftisol", linewidth=.5)
plt.plot(df_test.index.values, df_test.loc[:,"Placebo"] / df_test.loc[0,"Placebo"] * 100,
         marker ='d', color='black', label="Placebo", linewidth=.5)
plt.plot(df_test.index.values, df_test.loc[:,"Propriva"] / df_test.loc[0,"Propriva"] * 100,
         marker ='*', color='aqua', label="Propriva", linewidth=.5)
plt.plot(df_test.index.values, df_test.loc[:,"Ramicane"] / df_test.loc[0,"Ramicane"] * 100,
         marker ='p', color='brown', label="Ramicane", linewidth=.5)
plt.plot(df_test.index.values, df_test.loc[:,"Stelasyn"] / df_test.loc[0,"Stelasyn"] * 100,
         marker ='|', color='pink', label="Stelasyn", linewidth=.5)
plt.plot(df_test.index.values, df_test.loc[:,"Zoniferol"] / df_test.loc[0,"Zoniferol"] * 100,
         marker ='+', color='purple', label="Zoniferol", linewidth=.5)
# Format plot with title, axis labels, grid, and legend
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.grid()
lgnd = plt.legend(loc="lower left", fontsize=8)


# Save the Figure

# Show the Figure
plt.show()
```

```python
# Calculate the percent changes for each drug
# tumor_pct_change = 100* ((mean_tumor_volume.iloc[-1]- mean_tumor_volume.iloc[0])/mean_tumor_volume.iloc[0])           
# print(tumor_pct_change)

# tumor_pct_change_sem = 100* ((stand_err_tumor.iloc[-1]-stand_err_tumor.iloc[0])/stand_err_tumor.iloc[0])
# tumor_pct_change_sem
# # Display the data to confirm

# Drop intermediate Timepoints, keeping only first (0) and last (45)
start_end_df = df_format.drop([5,10,15,20,25,30,35,40], axis = 0)

# Add a calculated row for Percent Change
pct_response_df = start_end_df.append( (start_end_df.loc[45,:] - start_end_df.loc[0,:]) / start_end_df.loc[0,:] * 100, ignore_index=True)

# Drop rows for Timepoints 0 and 45, keeping only the Percent Change row
pct_response_df.drop([0,1], axis = 0, inplace = True)

# Transpose the DataFrame
pct_response_df = pct_response_df.transpose()

# Rename the Percent Change column
pct_response_df = pct_response_df.rename(columns={2:"Percent Change"})

# Preview DataFrame
pct_response_df.head()

# tumor_pct_change = all_mouse_data.pivot_table('Tumor Volume (mm3)', 'Drug', 'Timepoint', aggfunc=np.mean)
# avg_vol_change = (tumor_pct_change[max(tumor_pct_change.columns.values)]), tumor_pct_change[min(tumor_pct_change.columns.values)]/tumor_pct_change[min(tumor_pct_change.columns.values)]*100
# avg_vol_change


# 
```
